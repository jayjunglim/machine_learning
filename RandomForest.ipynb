## 랜덤포레스트

- bagging을 사용하는 대표적인 알고리즘. 배깅이란 여러개의 분류기를 만들어 마지막에 voting하여 최종 결정하는 알고리즘
- 또한,여러개의 샘플링을 통해 다양한 훈련용 데이터를 만듬
- 결정 트리(decision tree)를 기반으로 하고있음


![image](https://miro.medium.com/max/592/1*i0o8mjFfCn-uD79-F1Cqkw.png)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
def get_human_dataset():
    feature_name_df = pd.read_csv("./features.txt", sep ='\s+', header = None, names =['column_index','column_name'])
    feature_name = feature_name_df.iloc[:,1].values.tolist()
    
    X_train = pd.read_csv('./X_train.txt', sep ='\s+', names =feature_name)
    X_test = pd.read_csv('./X_test.txt', sep = '\s+', names = feature_name)
    y_train = pd.read_csv('./y_train.txt', sep = '\s+', header = None, names =['action'])
    y_test = pd.read_csv('./y_test.txt', sep = '\s+', header = None, names = ['action'])
    return X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = get_human_dataset()
```

RF를 이용한 정확도 산출

```python
rf_clf = RandomForestClassifier(random_state = 0)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print("정확도 : {0: .4f}".format(accuracy))
```

### gridserach를 이용한 하이퍼 파라미터 튜닝

단, 트리기반 앙상블 알고리즘들은 높은 정확도를 보여주는 대신 하이퍼 파라미터가 많아서 소요시간이 오래걸림
GBM, XG, LightGBM으로 넘어가면 속도가 느려짐

#### 주요 파라미터

1. n_estimator: 결정 트리의 갯수. default = 10
2. max_features: 결정 트리에 사용된 max_features. 최적의 분할을 위한 고려할 최대 feature 갯수
3. max_depth: 트리의 최대 깊이 규정
4. min_samples_leaf: 리프노드가 되기 위한 최소의 노드 수

#### 방법
params를 만든 뒤 n_estimator는 하나라 고정. 나머지를 먼저 설정 중요한 파라미터만 grid search로 탐색 후 고정하기 위함(시간소요를 줄여줌)

```python
params = {
    'n_estimators' : [100],
    'max_depth' : [6,8,10,12],
    'min_samples_leaf' : [8,12,16],
    'min_samples_split' : [8,16,24]
}
rf_clf = RandomForestClassifier(random_state = 0, n_jobs = -1)   #random_state: 시드설정, n_jobs: 아마 병렬처리 -1이면 모든 프로세스 활성화
grid_cv = GridSearchCV(rf_clf, param_grid = params, cv = 2, n_jobs = -1)   # param_grid: 검색할 모든 sequence, cv: Kfold 할수 (정수)
grid_cv.fit(X_train, y_train)

print("최적 하이퍼 파라미터 : ", grid_cv.best_params_)
print("최고 예측 정확도 : ", grid_cv.best_score_)
```

하이퍼 파라미터를 고정후 n_estimators를 변형시켜가며 확인

```python
params = {'n_estimators': [100,300,500]}
rf_clf_1 = RandomForestClassifier(max_depth=10, min_samples_leaf=8, min_samples_split=8, n_jobs=-1)
grid_cv = GridSearchCV(rf_clf_1, param_grid=params, cv = 3, n_jobs = -1)
grid_cv.fit(X_train,y_train)
print('최적 파라미터 : ', grid_cv.best_params_)
print('최고 예측 정확도 : ', grid_cv.best_score_)

best_model = grid_cv.best_estimator_
pred = best_model.predict(X_test)
acc = accuracy_score(y_test, pred)
print("정확도 : {0:.4f}".format(acc))
```

feature_importances_ 뽑아내기

```python
feature_importances = best_model.feature_importances_
ft_importances = pd.Series(feature_importances, index = X_train.columns)
ft_importances.iloc[:10]
```

 변수 중요도 시각화

```python
ft_top20 = ft_importances.sort_values(ascending = False)[:20]

plt.figure(figsize = (12,10))
plt.title("feature importances to 20")
sns.barplot(x = ft_top20, y = ft_top20.index)
plt.show()
```
